```python

from statsmodels.stats.outliers_influence import variance_inflation_factor

# Drop the target variable 'HousePrice' (ensure the case matches your DataFrame's column names)
housing_df_vif = housing_df.drop('HousePrice', axis=1)

# Ensure all columns are numeric
housing_df_vif = housing_df_vif.apply(pd.to_numeric, errors='coerce')

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = housing_df_vif.columns
vif_data["VIF"] = [variance_inflation_factor(housing_df_vif.values, i) for i in range(housing_df_vif.shape[1])]

print(vif_data)
               
              
```

```python
 #since latitude and longitude has high vif values so now removing  
```

```python
housing_df= housing_df.drop(['Latitude','Longitude'],axis=1)
```

```python
# Columns to convert to numeric
cols_to_convert = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup']
housing_df[cols_to_convert] = housing_df[cols_to_convert].apply(pd.to_numeric, errors='coerce')
housing_df['HousePrice'] = pd.to_numeric(housing_df['HousePrice'], errors='coerce')

# Define independent variables (X) and dependent variable (y)
X = sm.add_constant(housing_df.drop('HousePrice', axis=1))
y = housing_df['HousePrice']

# Fit the OLS model
model = sm.OLS(y, X).fit()

# Display the model summary
print(model.summary())

```

```python
residual= model.resid
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot(residual, kde=True)
plt.title('Residuals Distribution')
plt.subplot(1,2,2)
plt.scatter(model.predict(X),residual)
plt.axhline(y=0,color='red',linestyle='--')
plt.title("residual vs predicted")
plt.ylabel('Residual')
plt.show()
print("model coefficient:\n",model.params)
```

```python
#since the value are predicted to be at near 0 but the value not near so outliers are present 
#form the model coefficient ,we know that medinc is positive correlation with house pricealso avebedrms shows the most significiant influence to houseprice
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predicting with the model
prediction = model.predict(X)

# Printing the R^2 score and Mean Squared Error
print("R^2 score:", r2_score(y, prediction))
print("MSE:", mean_squared_error(y, prediction))
```

```python
# here In this model,54% of variance in the dependent variable with mean square error of 0.61 indicitaing model predictive performance
```

```python
#box plot to identify the outliers
import matplotlib.pyplot as plt

def plot_boxplots(df):
    fig, axes = plt.subplots(1, len(df.columns), figsize=(20, 5))
    for i, column_name in enumerate(df.columns):
        df.boxplot(column=column_name, ax=axes[i])
        axes[i].set_title(column_name)
    plt.tight_layout()
    plt.show()

# Call the function to plot boxplots
plot_boxplots(housing_df)
```

```python
#removing the ouliers
def remove_outliers(df):
    df_no_outliers = df.copy()
    for column_name in df.columns:
        Q1 = df[column_name].quantile(0.25)
        Q3 = df[column_name].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df_no_outliers[column_name] = df_no_outliers[column_name].apply(
            lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x
        )
    return df_no_outliers

# Apply the function to the DataFrame
housing_df_new = remove_outliers(housing_df)
```

```python
plot_boxplots(housing_df_new)
```

```python
#fit the linear regression
cols_to_convert = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup']
housing_df_new[cols_to_convert] = housing_df_new[cols_to_convert].apply(pd.to_numeric, errors='coerce')
housing_df_new['HousePrice'] = pd.to_numeric(housing_df_new['HousePrice'], errors='coerce')

# Define independent variables (X) and dependent variable (y)
X = sm.add_constant(housing_df_new.drop('HousePrice', axis=1))
y = housing_df_new['HousePrice']

# Fit the OLS model
model = sm.OLS(y, X).fit()


                  
```

```python
residuals = model.resid
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot(residuals, kde=True)
plt.title('Residuals Distribution')
plt.subplot(1,2,2)
plt.scatter(model.predict(X),residuals)
plt.axhline(y=0,color='red',linestyle='--')
plt.title("residuals vs predicted")
plt.xlabel('predicted values')
plt.ylabel('Residual')
plt.show()
print("model coefficient:\n",model.params)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Predicting with the model
prediction = model.predict(X)

# Printing the R^2 score and Mean Squared Error
print("R^2 score:", r2_score(y, prediction))
print("MSE:", mean_squared_error(y, prediction))
```
